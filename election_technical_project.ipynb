# Indian Elections 2019

**Introduction:**<br>
    
The Analysis here is based on the election of Lok Sabha for the year 2019, Members of Lok Sabha or the lower house 
of India's Parliament are elected by being voted upon by all adult citizens of India, from a set of candidates who stand in their respective constituencies. 
Every adult citizen of India can vote only in their constituency. Candidates who win the Lok Sabha elections are called 'Member of Parliament' and hold their seats 
for five years or until the body is dissolved by the President on the advice of the council of ministers.

# Data Cleaning

```python
# Clean Assets and LIABILITIES,EDUCATION,CRIMINAL CASES Column values

def data_clean(x):
    try:
        x=x.split('\n')[0]
        x=x.replace('Rs ','')
        x=x.replace(',','')
        x=x.strip()
        return int(x)
    except:
        x=0
        return int(x)

data['ASSETS']=data['ASSETS'].apply((data_clean))
data['LIABILITIES'] = data['LIABILITIES'].apply((data_clean))
data['EDUCATION']=data['EDUCATION'].str.replace('Post Graduate\n','Post Graduate')
data['CRIMINAL CASES']=data['CRIMINAL CASES'].str.replace('Not Available','0')

#to change data type of CRIMINAL CASES from string to float 
data=data.astype({'CRIMINAL CASES':float})
data.head()
```

```python
data.info()
```

```python
#To check null values
data.isnull().sum()
```

```python
#replacing null values with mode
data['GENDER'].fillna(data['GENDER'].mode()[0],inplace=True)
data['SYMBOL'].fillna(data['SYMBOL'].mode()[0],inplace=True)
data['CATEGORY'].fillna(data['CATEGORY'].mode()[0],inplace=True)
data['EDUCATION'].fillna(data['EDUCATION'].mode()[0],inplace=True)
data['CRIMINAL CASES'].fillna(data['CRIMINAL CASES'].mode()[0],inplace=True)

#replacing null values with mean
data['AGE'].fillna(data['AGE'].mean(),inplace=True)
data['ASSETS'].fillna(data['ASSETS'].mean(),inplace=True)
data['LIABILITIES'].fillna(data['LIABILITIES'].mean(),inplace=True)
```

```python
#To check null values
data.isnull().sum()
```

# columns of categorial data type

```python
#using fun_cat function we find Columns that are Categorical data type.
cat=[]
for col in data.columns:
    if data[col].nunique()<= 40:
        cat.append(col)

print("Columns that are Categorical data type are:",cat)

#cat list is created,having Categorical data type
```

```python
cat=data.loc[:,cat]
con=data.drop(cat,axis=1)
```

# Categorial columns

```python
#to check unique values & Number of unique values in each field
def fun_cat(col):
    if cat[col].isnull().sum()>0:
        cat[col].fillna(cat[col].mode()[0],inplace=True)
    print(f"{col}:Number of Unique Values are {cat[col].nunique()}")  
    
    print(f"{col}:Unique Values are {cat[col].unique()}")
    print("\n")   
```

```python
for col in cat.columns:
    fun_cat(col)
```

# Continues columns

```python
#to check Number of unique values in each field
def fun_num(i):
    if con[i].isnull().sum()>0:
        con[i].fillna(con[i].mean(),inplace=True)
    
    print(f"{i}:Number of Unique Values are {con[i].nunique()}") 
        
```

```python
for i in con.columns:
    fun_num(i)
```
